# Precision bar graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

# Sample data
data = {
    'model_name': ['EfficientNet', 'ResNet'],
    'Precision': [0.403429, 0.331755],
}

# Create a beautiful color palette
palette = sns.color_palette("muted")

# Create a bar graph for Precision
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='model_name', y='Precision', data=data, palette=palette)

# Display values above each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.4f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                textcoords='offset points')

plt.title('Precision Comparison')
plt.xlabel('Model Name')
plt.ylabel('Precision')

# Create a custom legend with colors and model names
legend_patches = [Patch(color=color, label=model_name) for color, model_name in zip(palette, data['model_name'])]
plt.legend(handles=legend_patches, title='Models',loc='upper center', bbox_to_anchor=(0.5, 1))
plt.show()
```
